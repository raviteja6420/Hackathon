## 11. In 2019 which batsman scored highest number of runs by hitting 6's and 4's?

```python
df_new = pd.merge(df_del, df_mat, how='left', left_on='match_id', right_on='id')
```

```python
bats_man = df_new[(df_new.season==2019) & ((df_new.batsman_runs==4) | (df_new.batsman_runs==6))].groupby('batsman')
```

```python
bats_man.batsman_runs.sum().sort_values(ascending=False)
```
